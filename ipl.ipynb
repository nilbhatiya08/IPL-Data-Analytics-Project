***Task=Find the win percentage of each team.***

```python
matches_played = pd.concat(
    [ipl['team1'], ipl['team2']]
).value_counts()

matches_won = ipl['winner'].value_counts()

win_percentage = pd.DataFrame({
    'Matches_Played': matches_played,
    'Matches_Won': matches_won
})

win_percentage['Win_Percentage'] = ((
    win_percentage['Matches_Won']
    / win_percentage['Matches_Played']
) * 100).astype(int)

win_percentage = win_percentage.sort_values(
    by='Matches_Played',
    ascending=False
)

print(win_percentage)
```

***Task=Calculate toss-win vs match-win percentage for every team.***

```python
ipl.head()
```

```python
toss_wins = ipl['toss_winner'].value_counts()

toss_match_wins = (
    ipl[ipl['toss_winner'] == ipl['winner']]
    ['winner']
    .value_counts()
)

team_stats = pd.DataFrame({
    'Toss_Wins': toss_wins,
    'Match_Wins_After_Toss': toss_match_wins
})

team_stats['Toss_to_Match_Win_%'] = (
    team_stats['Match_Wins_After_Toss']
    / team_stats['Toss_Wins']
    * 100
).round(2)

team_stats = team_stats.sort_values(
    by='Toss_to_Match_Win_%',
    ascending=False
)

print(team_stats)
```

***Task=Find the team with the highest average win margin (runs).***

```python
runs_wins=ipl[ipl['win_by_runs']>0]

avg_margin=runs_wins.groupby('winner')['win_by_runs'].mean()
print(avg_margin.sort_values(ascending=False))
```

***Task=Find the team with the highest average win margin (wickets).***

```python
wickets_wins=ipl[ipl['win_by_wickets']>0]
avg_margin=wickets_wins.groupby('winner')['win_by_wickets'].mean()
print(avg_margin.sort_values(ascending=False).astype(int))
```
